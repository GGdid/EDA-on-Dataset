Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

The data

```python
test = pd.read_csv('tested.csv')
test.head()
```

```python
test.info()
```

```python
test.shape 
```

```python
sns.barplot(x='Survived', y='Age',data = test,color = 'Green')
plt.show()
```

Seaborn library also has the palette attribute which you can use to give different colors to the bars, for example 'Rocket','Winter',etc. 

creating histrograms for your choice of column names

```python
sns.histplot(test['Sex'])
plt.show()
```

```python
sns.boxplot(x=test['Age'])
plt.show()
```

```python
sns.barplot(x='Age', y='Pclass', data=test)
plt.show()
```

```python
sns.violinplot(data=test,x='Sex',y = 'Pclass')
plt.show()
```

```python
sns.violinplot(data=test[['Sex', 'Age', 'Pclass']])
plt.show()
```

```python
sns.heatmap(test.isnull(),yticklabels = False,cbar = True, cmap = 'viridis')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=test)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Pclass',data=test,palette='magma')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Pclass',data=test,palette='rainbow')
```

```python
sns.displot(test['Age'].dropna(),kde=True,color='purple',bins=40)
```
